## Data Preprocessing

```python
columns_to_keep = ['price', 'surface_area', 'nb_rooms', 'equipment']

df_cleaned = df_cleaned[columns_to_keep]

df_cleaned.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
equipment_encoded = vectorizer.fit_transform(df_cleaned['equipment'])


equipment_df = pd.DataFrame(equipment_encoded.toarray(), columns=vectorizer.get_feature_names_out())


df_cleaned = pd.concat([df_cleaned.reset_index(drop=True), equipment_df], axis=1)


df_cleaned = df_cleaned.drop(columns=['equipment'])
```

```python
X = df_cleaned[['surface_area', 'nb_rooms', 'nb_baths', 'jardin privé', 'meublé', 'parking', 'parking exterieur', 'parking interieur', 'sécurité', 'terrasse', 'video interphone', 'vue sur la ville']]  # Choose relevant columns
y = df_cleaned['price']  


print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error, mean_squared_error

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
import xgboost as xgb

model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```
